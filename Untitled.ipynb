```python
import pandas as pd
import sqlite3
import json
pd.set_option('display.max_columns', None)

file_path = 'Business Analyst _ Assignment _ Dataset (1).xlsx' 
df = pd.read_excel(file_path)
conn = sqlite3.connect(':memory:')  

df.to_sql('Order_Details', conn, index=False, if_exists='replace')
```

```python
df['Parameters'] = df['Parameters'].fillna('{}')  # NaN filling
df_expanded = df['Parameters'].apply(lambda x: json.loads(x) if isinstance(x, str) else {}).apply(pd.Series) # parameter column Key Value as a column

df = df.drop(columns=['Parameters']).join(df_expanded)
```

```python
# Highest delta, Average time calculation
df['Install Time'] = pd.to_datetime(df['Install Time'])
df['Event Time'] = pd.to_datetime(df['Event Time'])

df['Install_to_Event_Hours'] = (df['Event Time'] - df['Install Time']).dt.total_seconds() / 3600


media_delta = df.groupby('Media Source')['Install_to_Event_Hours'].mean().reset_index()
media_delta.rename(columns={'Install_to_Event_Hours': 'Avg_Delta_Hours'}, inplace=True)


event_avg_time = df.groupby('Event Name')['Install_to_Event_Hours'].mean().reset_index()


print("Media Source with Highest Delta:")
print(media_delta.sort_values(by='Avg_Delta_Hours', ascending=False).head(1))

print("\nAverage Time from Install to Each Event:")
print(event_avg_time)
```

```python
# Avg time delta H
df['Install_to_Event_Hours'] = (df['Event Time'] - df['Install Time']).dt.total_seconds() / 3600

df = df.sort_values(by=['customer_id', 'Event Time'])


df['Cumulative_Avg_Time'] = df.groupby('customer_id')['Install_to_Event_Hours'].expanding().mean().reset_index(level=0, drop=True)

overall_cumulative_avg = df['Cumulative_Avg_Time'].mean()

print(f"Cumulative Average Time from Install to All Events Together: {overall_cumulative_avg:.2f} hours")
```

```python
# Avg time detla min.
df['Install_to_Event_Minutes'] = (df['Event Time'] - df['Install Time']).dt.total_seconds() / 60


df = df.sort_values(by=['customer_id', 'Event Time'])

df['Cumulative_Avg_Time_Minutes'] = df.groupby('customer_id')['Install_to_Event_Minutes'].expanding().mean().reset_index(level=0, drop=True)

overall_cumulative_avg_minutes = df['Cumulative_Avg_Time_Minutes'].mean()

print(f"Cumulative Average Time from Install to All Events Together: {overall_cumulative_avg_minutes:.2f} minutes")
```

```python
#Revenue by Channel
revenue_by_channel = df.groupby(['Media Source', 'Channel'])['af_revenue'].sum().reset_index()

most_revenue_channel = revenue_by_channel.sort_values(by='af_revenue', ascending=False).iloc[0]

print("Most Revenue-Driving Channel:")
print(most_revenue_channel)
```

```python
# revenue by media source and channel
quality_df = df[(df['is_core_customer'] == 1) & (df['is_switch_added'] == 1) & (df['has_coupon_code'] == 0)]

quality_revenue_by_channel = quality_df.groupby(['Media Source', 'Channel'])['af_revenue'].sum().reset_index()


merged_revenue = revenue_by_channel.merge(quality_revenue_by_channel, on=['Media Source', 'Channel'], how='left', suffixes=('_Total', '_Quality'))


merged_revenue['Quality_Revenue_Percentage'] = (merged_revenue['af_revenue_Quality'] / merged_revenue['af_revenue_Total']) * 100

merged_revenue = merged_revenue.sort_values(by='af_revenue_Total', ascending=False)
print(merged_revenue)
```

```python
# QOQ
import pandas as pd
import sqlite3


df["Event Time"] = pd.to_datetime(df["Event Time"])


df["Order_Year"] = df["Event Time"].dt.year
df["Order_Quarter"] = df["Event Time"].dt.quarter


conn = sqlite3.connect(":memory:")
df.to_sql("orders", conn, index=False, if_exists="replace")

# SQL Query for Quarter-on-Quarter Retention
query = """
WITH CustomerOrders AS (
    SELECT 
        customer_id,
        Order_Year,
        Order_Quarter
    FROM orders
    WHERE "Event Time" >= date('now', '-1 year')
),
RetentionData AS (
    SELECT
        c1.customer_id,
        c1.Order_Year,
        c1.Order_Quarter AS Current_Quarter,
        COUNT(DISTINCT c1.customer_id) AS Current_Customers,
        COUNT(DISTINCT c2.customer_id) AS Previous_Customers
    FROM CustomerOrders c1
    LEFT JOIN CustomerOrders c2
        ON c1.customer_id = c2.customer_id
        AND c2.Order_Year = c1.Order_Year
        AND c2.Order_Quarter = c1.Order_Quarter - 1
    GROUP BY c1.Order_Year, c1.Order_Quarter
),
QoQ_Customer AS (
    SELECT
        Order_Year,
        SUM(CASE WHEN Current_Quarter = 1 THEN Current_Customers END) AS Q1_Customers,
        SUM(CASE WHEN Current_Quarter = 2 THEN Current_Customers END) AS Q2_Customers,
        SUM(CASE WHEN Current_Quarter = 3 THEN Current_Customers END) AS Q3_Customers,
        SUM(CASE WHEN Current_Quarter = 4 THEN Current_Customers END) AS Q4_Customers,
        SUM(CASE WHEN Current_Quarter = 1 THEN Previous_Customers END) AS Q1_Previous_Customers,
        SUM(CASE WHEN Current_Quarter = 2 THEN Previous_Customers END) AS Q2_Previous_Customers,
        SUM(CASE WHEN Current_Quarter = 3 THEN Previous_Customers END) AS Q3_Previous_Customers,
        SUM(CASE WHEN Current_Quarter = 4 THEN Previous_Customers END) AS Q4_Previous_Customers
    FROM RetentionData
    GROUP BY Order_Year
    ORDER BY Order_Year
)
SELECT 
    Order_Year,
    COALESCE(Q1_Customers * 100 / NULLIF(Q1_Previous_Customers, 0), 0) AS Q1_Retention,
    COALESCE(Q2_Customers * 100 / NULLIF(Q2_Previous_Customers, 0), 0) AS Q2_Retention,
    COALESCE(Q3_Customers * 100 / NULLIF(Q3_Previous_Customers, 0), 0) AS Q3_Retention,
    COALESCE(Q4_Customers * 100 / NULLIF(Q4_Previous_Customers, 0), 0) AS Q4_Retention
FROM QoQ_Customer;
"""

df_retention = pd.read_sql(query, conn)

print(df_retention)
```

```python
query = """
WITH CampaignPerformance AS (
    SELECT 
        "Campaign ID" AS Campaign_ID, 
        "Media Source" AS Media_Source,
        COUNT(*) AS total_events,
        MIN("Event Time") AS campaign_start_date
    FROM orders
    WHERE "Campaign ID" IS NOT NULL
    GROUP BY "Campaign ID", "Media Source"
)
SELECT * FROM CampaignPerformance 
WHERE Campaign_ID IN (
    SELECT Campaign_ID FROM CampaignPerformance ORDER BY total_events DESC LIMIT 5
)
UNION ALL
SELECT * FROM CampaignPerformance 
WHERE Campaign_ID IN (
    SELECT Campaign_ID FROM CampaignPerformance ORDER BY total_events ASC LIMIT 5
)
ORDER BY total_events DESC;
"""


df_campaign_performance = pd.read_sql(query, conn)


print(df_campaign_performance)
```

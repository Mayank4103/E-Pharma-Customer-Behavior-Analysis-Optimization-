```python
import pandas as pd
import json

file_path = 'Business Analyst _ Assignment _ Dataset (1).xlsx' 
df = pd.read_excel(file_path)

df['Parameters'] = df['Parameters'].fillna('{}')


col_index = df.columns.get_loc('Parameters')  
after_params_cols = df.iloc[:, col_index + 1:].copy() 


df_expanded = df['Parameters'].apply(lambda x: json.loads(x) if isinstance(x, str) else {}).apply(pd.Series)


df = df.iloc[:, :col_index].join(df_expanded).join(after_params_cols)

print(df.head())
```
